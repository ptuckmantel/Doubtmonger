# Webscraping Desmoblog list of climate change denying institutions

Webscrape the list from the Desmoblog website. 
https://www.desmogblog.com/global-warming-denier-database

The goal is to then use google news search ' "[institute name]" "climate change" ' and get the first few pages.
    
This will allow us to get a dataset of climate change denying articles with the body of text, year, institution

```python
import requests
from bs4 import BeautifulSoup as bs
```

```python
url = "https://www.desmogblog.com/global-warming-denier-database"
page = requests.get(url)
soup = bs(page.content, 'html.parser')
```

```python
results = soup.find_all('h2', class_='views-field')
```

```python
# Check that we extract the part of the webpage containing the list of institutions
for j in results:
    print(j)
    print('\n')
```

```python
# Extract the institution names from the links
denier_institutions = []
for r in results:
    inst_name = r.find('a').contents[0]
    denier_institutions.append(inst_name)
```

```python
denier_institutions
```

```python
import pandas as pd

# Save to xlsx file for later use
df = pd.DataFrame({'denier institutions': denier_institutions})
df.to_excel('denier institutions.xlsx', index=False)
```
